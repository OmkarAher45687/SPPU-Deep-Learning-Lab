```python
import numpy as np  # Importing the NumPy library for numerical operations
import pandas as pd  # Importing the Pandas library for data manipulation
import seaborn as sns  # Importing Seaborn for data visualization
import matplotlib.pyplot as plt  # Importing Matplotlib for plotting graphs

# Importing necessary modules from Keras for building a deep learning model
from keras.models import Sequential  # Sequential model for building neural networks layer by layer
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D  # Layers for creating CNN architecture
from keras.utils import to_categorical  # Utility for converting labels to one-hot encoding
from sklearn.metrics import classification_report, confusion_matrix  # Metrics for model evaluation

# Load the training and testing data from CSV files
train_data = pd.read_csv('D:\\7th sem\\Deep Learning\\Dataset\\Fashion MNIST\\fashion-mnist_train\\fashion-mnist_train.csv')  # Load training data into a DataFrame
test_data = pd.read_csv('D:\\7th sem\\Deep Learning\\Dataset\\Fashion MNIST\\fashion-mnist_test\\fashion-mnist_test.csv')  # Load testing data into a DataFrame

# Extract features (pixel values) from the training data, skipping the first column (label)
X_train = train_data.iloc[:, 1:].values  # Select all rows and columns from the second column onward

# Extract labels (target values) from the training data (first column)
y_train = train_data.iloc[:, 0].values  # Select the first column as labels

# Extract features (pixel values) from the testing data, skipping the first column (label)
X_test = test_data.iloc[:, 1:].values  # Select all rows and columns from the second column onward

# Extract labels (target values) from the testing data (first column)
y_test = test_data.iloc[:, 0].values  # Select the first column as labels
```

```python
# Ensure the target variables (labels) are correctly one-hot encoded for training and testing
y_train = to_categorical(y_train, num_classes=10)  # Converts labels into a one-hot encoded format with 10 classes
y_test = to_categorical(y_test, num_classes=10)  # Converts labels into a one-hot encoded format with 10 classes

# Reshape the training and testing data and normalize pixel values
X_train = X_train.reshape(X_train.shape[0], 28, 28, 1).astype('float32') / 255.0  # Reshape to 28x28x1 and scale to [0, 1]
X_test = X_test.reshape(X_test.shape[0], 28, 28, 1).astype('float32') / 255.0  # Reshape to 28x28x1 and scale to [0, 1]

# Define the CNN model structure
model = Sequential()  # Create a Sequential model

# Add the first convolutional layer with 32 filters, 3x3 kernel, ReLU activation, and input shape (28, 28, 1)
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))

# Add a max pooling layer with a 2x2 pool size to reduce dimensionality
model.add(MaxPooling2D((2, 2)))

# Add the second convolutional layer with 64 filters, 3x3 kernel, and ReLU activation
model.add(Conv2D(64, (3, 3), activation='relu'))

# Add another max pooling layer with a 2x2 pool size
model.add(MaxPooling2D((2, 2)))

# Add the third convolutional layer with 64 filters, 3x3 kernel, and ReLU activation
model.add(Conv2D(64, (3, 3), activation='relu'))

# Flatten the output to convert the 3D output to 1D for the fully connected layer
model.add(Flatten())

# Add a fully connected (dense) layer with 64 units and ReLU activation
model.add(Dense(64, activation='relu'))

# Add a dropout layer to reduce overfitting, dropping 50% of the units
model.add(Dropout(0.5))

# Add the output layer with 10 units (one for each class) and softmax activation for multi-class classification
model.add(Dense(10, activation='softmax'))
```

```python
# Compile the model with the specified configuration
model.compile(
    optimizer='adam',               # Use the Adam optimizer, which adapts the learning rate for better convergence
    loss='categorical_crossentropy', # Use categorical crossentropy as the loss function for multi-class classification
    metrics=['accuracy']            # Track accuracy as a metric to evaluate the model's performance
)


# Train the model using the training data
history = model.fit(
    X_train,                # The input training data (features)
    y_train,                # The corresponding training labels (targets)
    epochs=4,               # Number of complete passes through the training dataset
    batch_size=32,          # Number of samples per gradient update (smaller batches use less memory but may be noisier)
    validation_split=0.2,   # Fraction of the training data (20%) to be used for validation during training
    validation_data=(X_test, y_test)  # The testing data used for validation after each epoch
)
```

```python
# Evaluate the trained model on the test dataset to determine its final performance
test_loss, test_acc = model.evaluate(
    X_test,  # Test data (features)
    y_test,  # Test labels (one-hot encoded)
    verbose=2  # Verbosity mode: 2 means one line per epoch is printed during evaluation
)

# Print the test accuracy in a formatted way
print(f'\nTest accuracy: {test_acc:.2f}')  # Display the accuracy as a percentage with 2 decimal places
```

```python
# Make predictions on the test data using the trained model
predictions = model.predict(X_test)  # The output will be a 2D array with probabilities for each class

# Convert the predicted probabilities to class labels by taking the index of the highest probability
y_pred = np.argmax(predictions, axis=1)  # Returns the index of the maximum value along the specified axis

# Convert the actual test labels from one-hot encoding to integers
y_true = np.argmax(y_test, axis=1)  # Returns the index of the maximum value along the specified axis for actual labels

# Print a detailed classification report showing precision, recall, and F1-score for each class
print("Classification Report:")
print(classification_report(
    y_true,  # True labels as integers
    y_pred,  # Predicted labels as integers
    target_names=['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']  # Names of the classes for better readability
))
```

```python
# Create a confusion matrix comparing true labels and predicted labels
cm = confusion_matrix(y_true, y_pred)

# Set up the figure size for better visualization
plt.figure(figsize=(10, 7))

# Plot the confusion matrix using a heatmap for visual representation
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot'],
            yticklabels=['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot'])

# Label the x-axis as 'Predicted Label'
plt.xlabel('Predicted Label')

# Label the y-axis as 'True Label'
plt.ylabel('True Label')

# Add a title to the plot
plt.title('Confusion Matrix')

# Display the heatmap
plt.show()
```

```python
# Importing the matplotlib library for plotting
import matplotlib.pyplot as plt

# Create a new figure with a specified size of 12 inches by 4 inches
plt.figure(figsize=(12, 4))

# Set up the first subplot for accuracy plotting
plt.subplot(1, 2, 1)  # 1 row, 2 columns, first subplot
# Plot the training accuracy values from the training history
plt.plot(history.history['accuracy'], label='Training Accuracy')
# Plot the validation accuracy values from the training history
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
# Set the title of the accuracy plot
plt.title('Model Accuracy')
# Label the x-axis as 'Epoch'
plt.xlabel('Epoch')
# Label the y-axis as 'Accuracy'
plt.ylabel('Accuracy')
# Show the legend in the lower right corner of the plot
plt.legend(loc='lower right')

# Set up the second subplot for loss plotting
plt.subplot(1, 2, 2)  # 1 row, 2 columns, second subplot
# Plot the training loss values from the training history
plt.plot(history.history['loss'], label='Training Loss')
# Plot the validation loss values from the training history
plt.plot(history.history['val_loss'], label='Validation Loss')
# Set the title of the loss plot
plt.title('Model Loss')
# Label the x-axis as 'Epoch'
plt.xlabel('Epoch')
# Label the y-axis as 'Loss'
plt.ylabel('Loss')
# Show the legend in the upper right corner of the plot
plt.legend(loc='upper right')

# Display the plots on the screen
plt.show()
```
