```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf 
from tensorflow import keras


from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.optimizers import SGD
```

```python
# Load your dataset from a CSV file
data = pd.read_csv('D:\\7th sem\\Deep Learning\\Dataset\\MNIST\\mnist_784_csv.csv')

# Display the first few rows of the dataset to understand its structure
print(data.head())

# Assuming the last column is the label and the rest are pixel values
x = data.iloc[:, :-1].values  # Features (pixel values)
y = data.iloc[:, -1].values    # Labels

# Reshape x to (num_samples, 28, 28)
x = x.reshape(-1, 28, 28)  # Reshape to (num_samples, height, width)

# Normalize the input data
x = x.astype('float32') / 255.0

# One-hot encode the labels
y = tf.keras.utils.to_categorical(y, num_classes=10)  # Assuming 10 classes (0-9)

# Split the dataset into training and testing sets
x_train, x_test = x[:60000], x[60000:]
y_train, y_test = y[:60000], y[60000:]
```

```python

# Build the model
model = Sequential([
    Flatten(input_shape=(28, 28)),  # Flatten the 28x28 images into a single vector
    Dense(50, activation='relu'),    # First hidden layer with 50 neurons
    Dense(650, activation='relu'),   # Second hidden layer with 650 neurons
    Dense(10, activation='softmax')   # Output layer for 10 classes (digits 0-9)
])

# Compile the model
model.compile(optimizer="sgd",
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Train the model
history = model.fit(x, y, epochs=10, batch_size=30, validation_split=0.2)
```

```python

# Evaluate the model on the test data
test_loss, test_acc = model.evaluate(x_test, y_test)
print(f'Test Loss: {test_loss}')
print(f'Test Accuracy: {test_acc}')

# Plot training loss and accuracy
plt.figure(figsize=(12, 4))
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Test Loss')
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(loc='upper left')

plt.figure(figsize=(12, 4))
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Test Accuracy')
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(loc='upper left')
plt.show()

# Display a random test image
import random
n = random.randint(0, len(x_test) - 1)

# Display the image
plt.imshow(x_test[n], cmap='gray')
plt.title(f"Label: {np.argmax(y_test[n])}")
plt.axis('off')  # Hide axis
plt.show()

# Visualize the first training image
plt.matshow(x_train[0], cmap='gray')
plt.title('First Training Image')
plt.axis('off')  # Hide axis
plt.show()
```
